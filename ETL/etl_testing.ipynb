# Imports & Setup

```python
from dotenv import load_dotenv
import os
import simfin as sf
import polars as pl
```

```python
load_dotenv()
```

```python
sf.set_api_key(os.getenv('API_KEY'))
sf.set_data_dir('data/raw')
```

# Data Ingestion

```python
pd_companies = sf.load_companies(market='us')
pd_share_prices = sf.load_shareprices(market='us')
```

```python
pl_companies = pl.DataFrame(pd_companies)
pl_share_prices = pl.DataFrame(pd_share_prices)
pl_companies.write_parquet('data/raw/parquet/companies.parquet')
pl_share_prices.write_parquet('data/raw/parquet/share_prices.parquet')
```
